```python
import pandas as pd
```

```python
df = pd.read_csv("./generated_responses_100.csv")
```

```python
df
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-4B")
```

```python
import numpy as np
tokens_human = [len(tokenizer.encode(text=text, add_special_tokens=False)) for text in df["response"].tolist()]
tokens_ai = [len(tokenizer.encode(text=text, add_special_tokens=False)) for text in df["generated_response"].tolist()]

print(np.mean(tokens_human))
print(np.mean(tokens_ai))
```

```python
import re
from nltk.util import ngrams
from collections import Counter

def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())  # lowercase + strip punctuation

ai_responses = df["generated_response"].tolist()
human_responses = df["ai_response"].tolist()

ai_str = " ".join(ai_responses)
human_str = " ".join(human_responses)

ai_words    = tokenize(ai_str)
human_words = tokenize(human_str)

unigrams_ai    = Counter(ngrams(ai_words, 1))
bigrams_ai    = Counter(ngrams(ai_words, 2))
trigrams_ai   = Counter(ngrams(ai_words, 3))
fourgrams_ai   = Counter(ngrams(ai_words, 4))

unigrams_human  = Counter(ngrams(human_words, 1))
bigrams_human  = Counter(ngrams(human_words, 2))
trigrams_human = Counter(ngrams(human_words, 3))
fourgrams_human = Counter(ngrams(human_words, 4))
```

```python
df_unigrams_ai = pd.DataFrame(
    {
    "words":unigrams_ai.keys(),
    "freaquency":unigrams_ai.values(),
    }
)
df_unigrams_human = pd.DataFrame(
    {
    "words":unigrams_human.keys(),
    "freaquency":unigrams_human.values(),
    }
)
df_bigrams_ai = pd.DataFrame(
    {
    "words":bigrams_ai.keys(),
    "freaquency":bigrams_ai.values(),
    }
)
df_bigrams_human = pd.DataFrame(
    {
    "words":bigrams_human.keys(),
    "freaquency":bigrams_human.values(),
    }
)
df_trigrams_ai = pd.DataFrame(
    {
    "words":trigrams_ai.keys(),
    "freaquency":trigrams_ai.values(),
    }
)
df_trigrams_human = pd.DataFrame(
    {
    "words":trigrams_human.keys(),
    "freaquency":trigrams_human.values(),
    }
)
df_fourgrams_ai = pd.DataFrame(
    {
    "words":fourgrams_ai.keys(),
    "freaquency":fourgrams_ai.values(),
    }
)
df_fourgrams_human = pd.DataFrame(
    {
    "words":fourgrams_human.keys(),
    "freaquency":fourgrams_human.values(),
    }
)
```

```python
df_unigrams_ai_sorted = df_unigrams_ai.sort_values("freaquency", ascending=False)
df_unigrams_human_sorted = df_unigrams_human.sort_values("freaquency", ascending=False)
df_bigrams_ai_sorted = df_bigrams_ai.sort_values("freaquency", ascending=False)
df_bigrams_human_sorted = df_bigrams_human.sort_values("freaquency", ascending=False)
df_trigrams_ai_sorted = df_trigrams_ai.sort_values("freaquency", ascending=False)
df_trigrams_human_sorted = df_trigrams_human.sort_values("freaquency", ascending=False)
df_fourgrams_ai_sorted = df_fourgrams_ai.sort_values("freaquency", ascending=False)
df_fourgrams_human_sorted = df_fourgrams_human.sort_values("freaquency", ascending=False)
```

```python
df_unigrams_ai_sorted.to_csv("./df_unigrams_ai_sorted_no_sw.csv", index=False)
df_unigrams_human_sorted.to_csv("./df_unigrams_human_sorted_no_sw.csv", index=False)
df_bigrams_ai_sorted.to_csv("./df_bigrams_ai_sorted.csv", index=False)
df_bigrams_human_sorted.to_csv("./df_bigrams_human_sorted.csv", index=False)
df_trigrams_ai_sorted.to_csv("./df_trigrams_ai_sorted.csv", index=False)
df_trigrams_human_sorted.to_csv("./df_trigrams_human_sorted.csv", index=False)
df_fourgrams_ai_sorted.to_csv("./df_fourgrams_ai_sorted.csv", index=False)
df_fourgrams_human_sorted.to_csv("./df_fourgrams_human_sorted.csv", index=False)
```
